```python
import pandas as pd
import os
from sklearn import tree

import numpy as np
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head(20)
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
data = df.drop("koi_disposition", axis=1)
feature_names = data.columns
data.head()
#selected_features = df[['names', 'of', 'selected', 'features', 'here']]
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
target = df["koi_disposition"]
target.value_counts()
print(target)
print(target.value_counts())
target_names = ["CONFIRMED", "FALSE POSITIVE", "CANDIDATE"]
print(target_names)
target.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
print(X_test)
print(y_test)
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
```

```python
 from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
#X_test_minmax = MinMaxScaler().fit(y_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
#y_train_minmax = y_minmax.transform(y_train)
#y_test_minmax = y_minmax.transform(y_test)
```

# Train the Model



```python
model1 = tree.DecisionTreeClassifier()
model1 = model1.fit(X_train, y_train)
model1.score(X_test, y_test)
#print(f"Training Data Score: {model1.score(X_train, y_train)}")
#print(f"Testing Data Score: {model1.score(X_test, y_test)}")
print(f"Training Data Score: {model1.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model1.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10, 50],
              'gamma': [0.0001, 0.0005, 0.001, 0.005]}
grid = GridSearchCV(model1, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
